```python
df= pd.read_csv('netflix.csv')
df.head()
```

```python
#Select TV Show released in year 2019
df[(df.release_year == 2019) & (df.type == 'TV Show')]
```

```python
#Select Movies between year 2018 to 2020
df[(df.release_year>2017) & (df.release_year<2021) & (df.type == 'Movie')]
```

```python
#Select shows aired only in United states
df[df.country.str.contains('United States') == True]
```

```python
#Select cast of the TV shows  telecasted in year 2019 and had rating R
df[(df.release_year == 2019) & (df.type == 'TV Show') & (df.rating == 'R')].cast
```
